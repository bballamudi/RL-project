# Example of how to use the cart pendulum environment

```python
# Import environment
from cartPendulum import cartPendulum

'''
The environment is initialized by creating an instance of the class cart pendulum
optional paramers are mass_cart, mass_pendulum, length pendulum, x0 (initial conditions) 
and gravity. The state of the environment is represented by an array with four vectors
[cart position, pendulum angle, cart velocity, pendulum angular velocity]
'''

env = cartPendulum()
env.step_size = 0.05 #The stepsize can also be set like this
```

## Environment simulation

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib auto

env.render() # Render the environment in real time


env.init([0, 0.1, 0, 0])
sampels = 100
y = np.zeros((sampels, 4))
y[0, :] = env.x


for t in range(sampels):
    s = env.state()
    env.step(np.random.random() - 0.5)  # Take random action at each time step
    y[t, :] = env.state()
    plt.pause(env.step_size)            # Enabels the plot to change in real time 

t = np.linspace(0, env.step_size*(sampels-1), sampels)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot(t,y)
plt.xlabel('Time [sec]')
plt.legend(['Cart position', 'Pendulum angle', 'Cart velocity', 'pendulum angular velocit'])
plt.show()
```
